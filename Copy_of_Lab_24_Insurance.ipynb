#3. Feature Selection + Linear Regression

Run a multiple linear regression machine learning algorithm which IVs are selected through SelectKbest f_regression. Select the best 3.

```python
# Calculate correlations between all columns (this computes the Pearson correlation coefficients between all pairs of numeric columns in the DataFrame)
correlations = df.corr()

# Identify potential features based on correlation with other columns
# Exclude highly correlated columns and non-numeric columns like 'ID'

# Filter out non-numeric columns and excluded columns
numeric_columns = df.select_dtypes(include=[np.number]).columns # selects only numeric columns
# filtered_features = [col for col in numeric_columns if col not in exclude_columns]
filtered_features = list(numeric_columns)

# Determine the target variable
# Find the column with the highest average correlation with other columns
avg_correlations = correlations.abs().mean() # computes mean of absolute correlation values for each column
target_variable = avg_correlations.idxmax() # identifies column w/highest average correlation (idxmax - index max)

# Define the feature matrix X and target variable y
X = df[filtered_features] # creates matrix x w/identified feature colulmns
y = df[target_variable] # sets target y to identified target column

# Display the identified features and target variable
print("\nIdentified Features:")
print(filtered_features)
print("\nTarget Variable:")
print(target_variable)
```

Target: charges

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

X = df.drop(['charges'], axis=1)
y = df['charges']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
df
```

```python
# Create a SelectKBest object with the f_regression scoring function
selector = SelectKBest(score_func=f_regression, k=3) #k=3 select 3

# Fit the selector to the training data and transform the features
X_train_selected = selector.fit_transform(X_train, y_train)
#Instead of having all features, X train and y train will not just have the best 5 columns

# Apply the same transformation to the test data
X_test_selected = selector.transform(X_test)
#y_test is not needed since we only use it in checking errors.

# Train a linear regression model using the selected features
model = LinearRegression()
model.fit(X_train_selected, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_selected)

y_pred
```

```python
# Print the indices of the selected features
selected_feature_indices = selector.get_support(indices=True)

# Get the indices of the selected features
print(f"Selected feature indices: {selected_feature_indices}")
```

```python
# Get the names of the selected features from the original DataFrame
selected_feature_names = X.columns[selected_feature_indices]

# Print the names of the selected features
print(f"Selected feature names: {selected_feature_names}")
```

#4. Evaluation

What is the MAPE, MAE, and RMSE? Check for both train and test data.


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Function to calculate Mean Absolute Percentage Error (MAPE)
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Train set evaluation
y_train_pred = model.predict(X_train_selected)

mape_train = mean_absolute_percentage_error(y_train, y_train_pred)
mae_train = mean_absolute_error(y_train, y_train_pred)
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))

print("Training Set Metrics with Selected Features:")
print("MAPE:", mape_train)
print("MAE:", mae_train)
print("RMSE:", rmse_train)

# Test set evaluation
y_test_pred = model.predict(X_test_selected)

mape_test = mean_absolute_percentage_error(y_test, y_test_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

print("\nTest Set Metrics with Selected Features:")
print("MAPE:", mape_test)
print("MAE:", mae_test)
print("RMSE:", rmse_test)
```

**Train Set:**

MAPE: Percentage of error at 43.58% in relation to actual values.

MAE: Average prediction error 4256.46 means predictions are off from actual by ~ 4256.46 dollars.

RMSE: Root of average square prediction error at 6,144.02 which relects the differences between predicted and actual value in units (charge dollars).

**Test Set:**

MAPE: Percentage of error at 46.91% relative to the actual values and is a little higher than the train set.

MAE: Average prediction error at 4274.43 means predictions are off from actual by ~4,274.43 dollars.

RMSE: Root of average square predition error at 5,997.79 which relects the differences between predicted and actual value in units (charge dollars).
