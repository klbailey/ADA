```python
from joblib import load

# Specify the path  'trained_model.joblib' in Colab
model_path = 'trained_model.joblib'

# Load the saved model
model = load(model_path)
```

###Create a DataFrame with the client information

```python
import pandas as pd

# Define the client information
client_data = {
    'age': [30, 27, 50],
    'sex': ['male', 'female', 'female'],
    'bmi': [28, 22, 24],
    'children': [0, 1, 2],
    'smoker': ['no', 'no', 'yes'],
    'region': ['southwest', 'southeast', 'southwest']
}

# Create a DataFrame from the client data
client_df = pd.DataFrame(client_data)
```

```python
client_df #(0=Sheldon Cooper, 1=Lily Potter, 2=Morticia Addams)
```

##Datapreprocessing

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
labelenc = LabelEncoder()

# Apply LabelEncoder to categorical columns and print classes in for loop
for col in ['sex', 'smoker', 'region']:
    client_df[col] = labelenc.fit_transform(client_df[col])
    print(f"Classes for '{col}': {labelenc.classes_}")

print()

# Display the updated DataFrame
client_df

# female=0, male=0; not smoker=0, smoker=1; region 0=southeast, 1=southwest (alphabetically)
```

##Make Predictions

Prepare selected features

```python
# selected_feature_indices is defined from previous feature selection step
selected_feature_indices = [0, 2, 4]  # Determined in Lab 24

# Select only the relevant features based on selected_feature_indices
client_df_selected = client_df.iloc[:, selected_feature_indices]
```

Predictions

```python
# Make predictions on client data
client_predictions = model.predict(client_df_selected)

# Print predictions
print("Predicted Charges for Clients:")
for prediction in client_predictions:
    print(prediction)
```
