## Description

The dataset provided is from the Centers for Medicaid and Medicare Services (CMS)and is known as the Payroll Based Journal (PBJ) Daily Nurse Staffing dataset. It offers detailed daily staffing records for registered nursing homes throughout the United States and is updated on a quarterly basis. The dataset for this analysis covers the most recent quarter, 2024Q1. The PBJ dataset is essential for analyzing staffing patterns, understanding the distribution of contractor versus employee hours, and assessing overall staffing efficiency in nursing homes. It provides valuable insights into how staffing levels are managed and where improvements may be needed to optimize resource allocation and meet the operational needs of long-term care facilities.<br><br>

Dataset Name: Payroll Based Journal Daily Nurse Staffing<br><br>

Source: Payroll Based Journal (PBJ) Daily Nurse Staffing Staffing [link](https://data.cms.gov/quality-of-care/payroll-based-journal-daily-nurse-staffing/data)

## Objective

The primary objective is to evaluate the contractor hours across various healthcare roles and develop a strategic plan to optimize the use of contractors, enhance workforce planning, and improve operational efficiency. The analysis aims to identify high and low areas of contractor usage, understand the implications for staffing, and propose actionable strategies to address identified needs.

###Outliers

---

The standard deviation (std) values suggest variability in hours worked across the roles. For example Hrs_CNA_ctr has a large standard deviation of 3.255127e+01 which is 32.55127 reflecting significant differences in contractor CNA hours across facilities.

```python
# Convert Hrs_CNA_ctr to millions for readability
non_zero_data['Hrs_CNA_ctr_millions'] = non_zero_data['Hrs_CNA_ctr'] / 1e6

# Define a threshold for outliers
threshold = non_zero_data['Hrs_CNA_ctr_millions'].quantile(0.95)  # Top 5% as an example
outliers = non_zero_data[non_zero_data['Hrs_CNA_ctr_millions'] > threshold]

# Print to verify outliers
print("Threshold:", threshold)
print("Number of Outliers:", len(outliers))

# Plot the data
plt.figure(figsize=(12, 8))
plt.scatter(non_zero_data.index, non_zero_data['Hrs_CNA_ctr_millions'], color='skyblue', label='Normal Data', alpha=0.6)
plt.scatter(outliers.index, outliers['Hrs_CNA_ctr_millions'], color='gold', label='Outliers', edgecolor='black')

plt.title('Contract Hours for CNA with Outliers Highlighted')
plt.xlabel('Index')
plt.ylabel('Contract Hours (in Millions)')
plt.legend()
plt.grid(True)

plt.show()
```

---

In the dataset, the threshold for identifying outliers in the contractor hours for CNA (Hrs_CNA_ctr) was set to approximately 0.00010933 million hours (or 109.33 hours). This threshold identifies the top 5% of data points as outliers, with a total of 28,631 entries exceeding this value.

Given that these outliers represent a substantial portion of the data and reflect real-world variations in staffing hours, it is crucial to include them in our analysis. Excluding these outliers would result in the loss of valuable information about extreme but realistic staffing scenarios. Therefore, we will retain these outliers to ensure our analysis accurately represents the true range of contractor hours.

##Exploratory Data Analysis (EDA)

###Histogram

```python
role_names = {
    'Hrs_CNA_ctr': 'Contract Hours for CNA',
    'Hrs_RN_ctr': 'Contract Hours for RN',
    'Hrs_NAtrn_ctr': 'Contract Hours for Nurse Aide in Training'
}

# Define the columns to plot
columns_to_plot = ['Hrs_CNA_ctr', 'Hrs_RN_ctr', 'Hrs_NAtrn_ctr']

# Plot histograms for each selected column
plt.figure(figsize=(10, 8))

for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(3, 1, i)  # 3 rows, 1 column, i-th subplot
    plt.hist(non_zero_data[column], bins=50, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {role_names[column]}')
    plt.xlabel('Hours')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

From the standard deviation completed prior, we look at low medium and high standard deviated roles for the plot.

- The contract CNA hours are highly variable with many facilities reporting lower hours, but many also reporting very high hours. The long right tail suggests that a small number of facilities have exceptionally high contract hours compared to the majority. This may indicate staffing needs for CNAs vary between facilities and some rely heavily on them.
- The contract RN hours are more concentrated when compared to CNAs. There is variability, but is more centralized around the lower end. The shorter right tail suggests fewer extreme high values and they are generally contracted in  more consistent amounts.
- The contract NA in Training hours are generally low with fewer facilities reporting high contract hours. The distribution is relatively uniform without significant outliers. This suggests the position is less variable in staffing needs or that facilities maintain a general consistent level of contract hours.

###Analyze Distribution by Role

```python
role_names = {
    'Hrs_RNDON_ctr': 'Contract Hours for RN Director of Nursing',
    'Hrs_RNadmin_ctr': 'Contract Hours for RN with Administrative Duties',
    'Hrs_RN_ctr': 'Contract Hours for RN',
    'Hrs_LPNadmin_ctr': 'Contract Hours for LPN with Administrative Duties',
    'Hrs_LPN_ctr': 'Contract Hours for LPN',
    'Hrs_CNA_ctr': 'Contract Hours for CNA',
    'Hrs_NAtrn_ctr': 'Contract Hours for Nurse Aide in Training',
    'Hrs_MedAide_ctr': 'Contract Hours for Med Aide/Technician'
}
```

```python
non_zero_data = df[(df[role_names.keys()] > 0).any(axis=1)]

# Aggregate Contract Hours by Role
contractor_hours_by_role = non_zero_data[role_names.keys()].sum()

# Create a DataFrame for plotting
contractor_hours_df = contractor_hours_by_role.reset_index()
contractor_hours_df.columns = ['Role', 'Total Contractor Hours']

# Map Role Names to Descriptive Labels
contractor_hours_df['Role'] = contractor_hours_df['Role'].map(role_names)

# Convert to Millions for Readability
contractor_hours_df['Total Contractor Hours (Millions)'] = contractor_hours_df['Total Contractor Hours'] / 1e6

# Plot Using Matplotlib
plt.figure(figsize=(12, 8))
bars = plt.barh(contractor_hours_df['Role'], contractor_hours_df['Total Contractor Hours (Millions)'], color='skyblue')

# Add Data Labels
for bar in bars:
    plt.text(
        bar.get_width() + 0.1,
        bar.get_y() + bar.get_height() / 2,
        f'{bar.get_width():,.1f}',
        va='center'
    )

# Update Layout to Include Grid
plt.xlabel('Total Contractor Hours (Millions)')
plt.ylabel('Role')
plt.title('Total Contractor Hours by Role (in Millions)')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save and Show the Plot
plt.savefig('/content/contractor_hours_plot_matplotlib.png')
plt.show()
```

**Analysis Based on Role Distribution**

Top Highest Contractor Hours:
- CNA (17.3 million hours) indicates a significant reliance on contractor CNAs. This may be due to fluctuating demand or staffing shortages.
- LPN (8.6 million hours) indicates a strong use of contractor LPNS possibly also due to fluctuating demand or staffing shortages.

Lowest Contractor Hours:
- RN Director of Nursing (0.1 million hours) implies minimal use for a high-level role. It suggests that nursing homes prefer to employ RN Directors of Nursing directly rather than using contractors. This may be due to the critical nature of the role and the need for continuity and consistency in leadership.
- LPN with Administrative Duties (0.1 million hours) implies less reliance on this position. This might be due to the need for these roles to be filled by permanent employees who are more familiar with the facility's operations.
- Nurses Aide in Training (0.01 million hours) is very low contacted hours. It implies that the role is predominantly filled by permanent staff rather than contractors. This could be because training roles require a stable and consistent presence, which is less feasible with contractors.

The remaining roles imply intermediate usage. This suggests a balanced approach to staffing these positions, where facilities use a mix of employees and contractors based on specific needs and availability.

## Recommendation Based on Role Distribution

- Focus on CNA Staffing Solutions
  - Focus on offering scalable CNA staffing solutions
  - Possibly provide flexible contracting options to meet fluctuating demands and create a pool of high-quality CNAs readily available to fill staffing shortages in nursing homes
- Evaluate Use of LPNs and Enhance Offerings
  - With (LPNs) at 8.6 million contractor hours, we could expand recruitment efforts for LPNs and offer competitive packages to retain these professionals
  - Emphasizing specialized LPN training and continuing education could increase appeal to nursing homes that are looking for skilled and adaptable LPN contractors
- Offer Specialized Solutions for Administrative Roles
  - Despite lower contractor hours for LPNs with administrative duties and RN Directors of Nursing, there may still be opportunities for short-term or interim staffing during transitions or vacancies
  - Developing niche contractor services for these specialized roles can help fill gaps in leadership or administration without a long-term commitment from facilities
- Promote flexibility for Intermediate Roles
  - For roles with intermediate usage, offering flexible contracting models could help nursing homes balance their use of permanent employees and contractors
  - We could focus on creating hybrid staffing models where facilities can adjust their workforce based on fluctuating patient volumes and seasonal demand
- Target Training and Development Services
  - While Nurses Aide in Training has low contractor usage, this could be an opportunity to partner with nursing homes for training programs
  - The company could provide contractor CNAs who double as trainers or facilitators, helping the facility develop its own staff while addressing immediate staffing needs
